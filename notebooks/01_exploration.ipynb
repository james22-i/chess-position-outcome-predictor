Core Idea: 
Imagine a complicated midgame position, there are many layers of ongoing attacks, defences, future threats and layers of protection.  
The hope is to quantify these variables and determine a relatively simple function which could indicate the likely outcome of the match.  
Potential variables come to mind: (layers of protection per piece, attacking moves per piece, current pieces under attack, pieces that could be put under attack)

We use the python-chess and the standardised FEN notation to make life easier.  
An example of FEN: r1b1k2N/ppp1q1pp/2n2n2/3p4/2B1P3/8/PPPP1bPP/RNBQ1K1R w q d6 0 1

```python
import chess
fen = "r1b1k2N/ppp1q1pp/2n2n2/3p4/2B1P3/8/PPPP1bPP/RNBQ1K1R w q d6 0 1"
print(chess.Board(fen))
```

The 3 first variables determined are: Position mobility, centrality and connection.<br>
Centrality is weighted on whether piece are on the inner most four square or the ones surrounding them.<br>
Connection is an idea from graph theory, the idea being that a protected and unified position would relate opposity to disconnections.<br>
For Mobility, different pieces potential moves are weighted differently, more weighting for a rook or queen, but less for a king. Would also be a good idea to investigate 2-stage mobility

In most midgame positions for higher-rated players the game is quite even, and training the regression model on the features based on the final outcome is misleading.
